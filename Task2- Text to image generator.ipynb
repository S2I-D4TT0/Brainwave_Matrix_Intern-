```python
from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt

import torch

# Load the pre-trained Stable Diffusion model from Hugging Face
model_id = "CompVis/stable-diffusion-v1-4"
device = "cuda" if torch.cuda.is_available() else "cpu"

# Load the pipeline with GPU support
pipeline = StableDiffusionPipeline.from_pretrained(model_id).to(device)

# Define the text prompt
prompt = "A student doing his project"

# Generate the image from the text prompt
image = pipeline(prompt).images[0]

# Save the generated image
image.save("generated_image.png")

# Show the generated image (optional)
plt.imshow(image)
plt.axis('off')
plt.show()
```
